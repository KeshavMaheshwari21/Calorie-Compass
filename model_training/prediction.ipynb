<h3>Importing Libraries :-</h3>

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor  
import pickle
```

```python
df = pd.read_csv("calories.csv") # loading the calorie data
df.head()
```

```python
df2 = pd.read_csv("exercise.csv") # loading the exercise data
df2.head()
```

```python
df3  = pd.concat((df,df2),axis=1) # concating the both dataset calorie and exercise
```

```python
df3.head()
```

```python
df3.info()
```

```python
df3.drop("User_ID",inplace=True,axis=1) # deleting the User_ID column from the dataframe
```

```python
df3.head()
```

```python
gender_mapping = {
    'female': 0,
    'male': 1
}
```

```python
df3['Gender'] = df3['Gender'].map(gender_mapping) # changing the catgorical data into numerical data through map function
```

```python
df3.head()
```

```python
df3.to_csv('data.csv',index=False) # saving the new dataframe in a csv file
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
x = df.drop("Calories",axis=1)
y = df.Calories
```

```python
# Splitting the data in training and testing with the help of train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# making objects of the model to train
lr_model = LinearRegression()
dt_model = DecisionTreeRegressor()
rf_model = RandomForestRegressor()
```

```python
#  training the model on the training data
lr_model.fit(X_train, y_train)
dt_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
```

```python
# creating the function for checking the training and testing score
def check_scores(model, model_name):
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    print(f"{model_name} Training Score: {train_score:.4f}")
    print(f"{model_name} Testing Score: {test_score:.4f}\n")
```

```python
# checking the training and testing score
check_scores(lr_model, "Linear Regression")
check_scores(dt_model, "Decision Tree Regressor")
check_scores(rf_model, "Random Forest Regressor")
```
